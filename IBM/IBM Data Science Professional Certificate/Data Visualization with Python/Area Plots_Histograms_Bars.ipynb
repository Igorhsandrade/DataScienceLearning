# Area Plots, Histograms, and Bar Chart

```python
import numpy as np 
import pandas as pd
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
```

```python
df_can.set_index('Country', inplace=True)
df_can.index.name = None
df_can.head(5)
```

## Area Plots

```python
years = list(map(str, range(1980,2014)))
df_top_5 = df_can.sort_values(by="Total", ascending=False, axis=0).head(5)[years].transpose()
df_top_5
```

```python
df_top_5.index = df_top_5.index.map(int)
```

### Plotting with scripting layer 

Use the plt and add more elements by caling different methods

```python
df_top_5.plot(kind="area", alpha=0.25, stacked=False, figsize=(20,20))
plt.title("Immigration Trend of Top 5 Countries")
plt.ylabel("Number of Immigrants")
plt.xlabel("Years")
plt.show()
```

### Ploting with artist layer - Axes instance

Use an Axes instance for the plot and store it in a variable 

```python
ax = df_top_5.plot(kind="area", alpha=0.35, figsize=(20,10))
ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

## Countries that contributed the last to immigration

```python
#Creating a stacked area with scripting
df_least_5 = df_can.sort_values(by="Total", ascending=True, axis=0).head(5)[years].transpose()
df_least_5
```

```python
df_least_5.plot(kind="area", alpha=0.45, figsize=(20,10))

plt.title('Immigration Trend of Least 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
ax = df_least_5.plot(kind='area', alpha=0.55, stacked=False, figsize=(20, 10))
    
ax.set_title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

## Histograms

```python
df_can['2013'].head()
```

```python
count, bin_edges = np.histogram(df_can['2013'])
print(f'Frequencies: {count} \nBin ranges (default = 10 bins): {bin_edges}')
```

```python
df_can['2013'].plot(kind='hist', figsize=(8,5))

plt.title('Histogram of Immigration from 195 countries in 2013')
plt.ylabel('Number of countries')
plt.xlabel('Number of Immigrants')

plt.show()
```

Analysing the distribution of Denmark, Norway, and Sweden for 1980-2013

```python
df_can.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
df_t = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.plot.hist()

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

Improving the aesthetics

```python
count, bin_edges = np.histogram(df_t, 15)

df_t.plot(kind= "hist",
          figsize= (10,6),
          bins= 15,
          alpha= 0.6,
          xticks= bin_edges,
          color= ['coral', 'darkslateblue', 'mediumseagreen']
        )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges[0] - 10   #  first bin value is 31.0, adding buffer of 10 for aesthetic purposes 
xmax = bin_edges[-1] + 10  #  last bin value is 308.0, adding buffer of 10 for aesthetic purposes

# stacked Histogram
df_t.plot(kind='hist',
          figsize=(10, 6), 
          bins=15,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'],
          stacked=True,
          xlim=(xmin, xmax)
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants') 

plt.show()
```

```python
df_gab = df_can.loc[['Greece','Albania','Bulgaria'], years].transpose()
df_gab
```

```python
cout, bin_edges = np.histogram(df_gab, 15)

df_gab.plot(kind= "hist",
            figsize= (10,6),
            bins= 15,
            alpha= 0.35,
            xticks= bin_edges,
            color=['coral', 'darkslateblue', 'mediumseagreen'])

plt.title('Histogram of Immigration from Greece, Albania, and Bulgaria from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

## Bar Charts

```python
df_iceland = df_can.loc['Iceland', years]
df_iceland.head()
```

```python
df_iceland.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Year') 
plt.ylabel('Number of immigrants') 
plt.title('Icelandic immigrants to Canada from 1980 to 2013') 

plt.show()
```

```python
df_iceland.plot(kind='bar', figsize=(10, 6), rot=90) 

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

# Annotate arrow
plt.annotate('', 
             xy=(32, 70),  
             xytext=(28, 20),  
             xycoords='data',  
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
             )

plt.annotate('2008 - 2011 Financial Crisis', 
             xy=(28, 30),  
             rotation=72.5, 
             va='bottom',  
             ha='left',  
             )
plt.show()
```

```python
df_top_15 = df_can.sort_values(by="Total", ascending=False, axis=0).head(15)['Total']
df_top_15
```

```python
df_top_15.plot(kind='bar',
               figsize=(10,6))
```

```python
df_top_15.plot(kind='barh',
               figsize=(12,12))

plt.xlabel('Number of Immigrants')
plt.title('Top 15 Conuntries Contributing to the Immigration to Canada between 1980 - 2013')

for index, value in enumerate(df_top_15):
    label = format(int(value),',')
    plt.annotate(label, xy=(value - 47000, index - 0.10), color='white')
```
