### Importing libraries

```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

### Importing data

```python
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv'
df_can = pd.read_csv(URL)
```

```python
df_can.head()
```

```python
df_can.set_index('Country', inplace=True)
df_can.index.name = None
df_can.head()
```

```python
df_can.shape
```

```python
years = list(map(str, range(1980,2014)))
```

## Pie Charts

```python
df_continents = df_can.groupby('Continent').sum()
df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                            figsize=(5,6),
                            autopct='1.1f%%',
                            startangle=90,
                            shadow=True
                           )

plt.title('Immigration to Canada by Continent [1980-2013]')
plt.axis('equal')
plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

#### Improving the vis

```python
colors_list = ['#66D9EF', '#8BC34A', '#9C27B0', '#3F51B5', '#03A9F4', '#FF9800']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]
df_continents['Total'].plot(kind='pie',
                            figsize=(10,6),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                            labels=None,
                            pctdistance=1.12,
                            colors=colors_list,
                            explode=explode_list
                           )

plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12, fontsize=15)
plt.axis('equal')
plt.legend(labels=df_continents.index, loc='upper left', fontsize=7)

plt.show()
```

```python
df_continents_2013 = df_continents['2013']
explode_list = [0, 0, 0, 0.1, 0.1, 0.2]
df_continents_2013.plot(kind='pie',
                            figsize=(15,6),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                            labels=None,
                            pctdistance=1.12,
                            colors=colors_list,
                            explode=explode_list
                           )

plt.title('Immigration to Canada by Continent in 2013', y=1.12)
plt.axis('equal')
plt.legend(labels=df_continents.index, loc='upper left')

plt.show()                        
```

```python
df_top_15 = df_can.sort_values(by='Total', ascending=False, axis=0).head(15)
df_top_15
```

```python
decades = [1980, 1990, 2000]
df_decades_sum = pd.DataFrame()
for decade in decades:
    decade_years_range = list(map(str,range(decade, decade+10)))
    s_decade = df_top_15.loc[:, decade_years_range].sum(axis=1)
    df_decades_sum[f'{decade}s'] = s_decade
df_decades_sum
```

```python
df_decades_sum.describe()
```

```python
df_decades_sum.plot(kind='box', figsize=(10,6))
plt.title('Immigration from top 15 countries for decades 80s, 90s and 2000s')
plt.show()
```

### Scatter Plots 

```python
df_total = pd.DataFrame(df_can[years].sum(axis=0))
df_total.index = map(int, df_total.index)
df_total.reset_index(inplace=True)
df_total.columns = ['year', 'total']
df_total.head()
```

```python
df_total.plot(kind='scatter', x='year', y='total', figsize=(10,6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
x = df_total['year']
y = df_total['total']
fit = np.polyfit(x, y, deg=1)

fit
```

```python
df_total.plot(kind='scatter', x='year', y='total', figsize=(10,6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.plot(x, fit[0] * x + fit[1], color='red')
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()

# print out the line of best fit
'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
```

```python
df_countries = df_can.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
df_total = pd.DataFrame(df_countries[years].sum(axis=0))
df_total.reset_index(inplace=True)
df_total.columns = ['year','total']
df_total['year'] = df_total['year'].astype(int)
df_total
```

### Bubble Plots

```python
df_can_t = df_can[years].transpose()

df_can_t.index = map(int, df_can_t.index)
df_can_t.index.name = 'Year'
df_can_t.reset_index(inplace=True)
df_can_t.head()
```

```python
norm_brazil = (df_can_t['Brazil'] - df_can_t['Brazil'].min()) / (df_can_t['Brazil'].max() - df_can_t['Brazil'].min())
norm_argentina = (df_can_t['Argentina'] - df_can_t['Argentina'].min()) / (df_can_t['Argentina'].max() - df_can_t['Argentina'].min())
```

```python
ax0 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='Brazil',
                    figsize=(14,8),
                    alpha=0.5,
                    color='green',
                    s = norm_brazil * 2000 + 10,
                    xlim=(1975,2015)
                   )
ax1 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='Argentina',
                    alpha=0.5,
                    color="blue",
                    s=norm_argentina * 2000 + 10,
                    ax=ax0
                    )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 to 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```

```python
norm_china = (df_can_t['China'] - df_can_t['China'].min()) / (df_can_t['China'].max() - df_can_t['China'].min())
norm_india = (df_can_t['India'] - df_can_t['India'].min()) / (df_can_t['India'].max() - df_can_t['India'].min())
```

```python
ax0 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='China',
                    figsize=(14,8),
                    alpha=0.5,
                    color='red',
                    s = norm_china * 2000 + 10,
                    xlim=(1975,2015)
                   )

ax1 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='India',
                    figsize=(14,8),
                    alpha=0.5,
                    color='orange',
                    s = norm_india * 2000 + 10,
                    ax= ax0             
                   )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from China and India from 1980 to 2013')
ax0.legend(['China', 'India'], loc='upper left', fontsize='x-large')                    
```
