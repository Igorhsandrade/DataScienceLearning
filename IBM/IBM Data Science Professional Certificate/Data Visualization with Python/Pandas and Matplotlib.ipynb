```python
df_can = pd.read_excel('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx',
                       sheet_name= "Canada by Citizenship",
                       skiprows=range(20),
                       skipfooter= 2,
                       engine='openpyxl')
```

```python
df_can.head()
```

```python
df_can.columns
```

```python
df_can.drop(['AREA','REG','DEV','Type','Coverage'], inplace=True, axis=1)
df_can.head()
```

```python
df_can.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'}, inplace=True)
df_can.head(2)
```

```python
df_can.dtypes
```

```python
df_can['Total'] = df_can.select_dtypes(include='number').sum(axis=1)
df_can['Total']
```

```python
df_can.isnull().sum()
```

```python
df_can.describe()
```

```python
df_can[['Country',2000,2001,2002,2003,2004,2005]]
```

```python
df_can.set_index('Country', inplace=True)
df_can.index.name = None
```

```python
df_can.columns = list(map(str, df_can.columns))
```

### Filtering by Country(index)

```python
years = list(map(str, range(1980,2014)))
japan_data = df_can.loc['Japan',years]
```

### Filtering by Continent and Region

* Continent: Africa
* Region: Southern Africa


```python
filtering_condition = (df_can['Continent'] == 'Africa') & (df_can['Region'] == 'Southern Africa')
df_can[filtering_condition]
```

### Finding top 3 countries that contributes the most to immigration in the year 2013

```python
df_can.sort_values(by='2013',ascending=False, axis=0, inplace=True)
top_3 = df_can.head(3)
top_3
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
japan_data.plot(kind="line")
plt.title('Immigration from Japan')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
japan_data.index = japan_data.index.map(int)
japan_data.plot(kind='line')
plt.title('Immigration from Japan')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.text(2000, 1100, 'Decrease')
plt.show()
```

### Comparing the number of immigrants from China and India

```python
china_and_india_data = df_can.loc[['China','India'],years]
china_and_india_data.head()
```

```python
china_and_india_data = china_and_india_data.transpose()
```

```python
china_and_india_data.plot(kind='line')
plt.title('Immigration from China & India')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```

```python
top_5 = df_can.sort_values(by="Total", ascending=False, axis=0).head(5)
top_5
```

```python
top_5 = top_5[years]
```

```python
top_5 = top_5.transpose()
```

```python
top_5.plot()
plt.title('Immigration from top 5 contributors')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```
